```python
board = [' '] * 9

def print_board():
    for i in range(0, 9, 3):
        print(board[i], "|", board[i+1], "|", board[i+2])
    print()

def check_winner(player):
    win = [(0,1,2),(3,4,5),(6,7,8),
           (0,3,6),(1,4,7),(2,5,8),
           (0,4,8),(2,4,6)]
    return any(board[a]==board[b]==board[c]==player for a,b,c in win)

def is_draw():
    return ' ' not in board

# Game loop (2 players)
player = 'X'

while True:
    print_board()
    pos = int(input(f"Player {player}, enter position (0-8): "))

    if board[pos] != ' ':
        print("Invalid move")
        continue

    board[pos] = player

    if check_winner(player):
        print_board()
        print(f"Player {player} wins!")
        break

    if is_draw():
        print_board()
        print("It's a draw!")
        break

    # switch player
    player = 'O' if player == 'X' else 'X'
```
