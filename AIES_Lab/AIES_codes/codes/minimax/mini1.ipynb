```python
board = [' '] * 9

def print_board():
    for i in range(0, 9, 3):
        print(board[i], "|", board[i+1], "|", board[i+2])
    print()

def check_winner(player):
    win = [(0,1,2),(3,4,5),(6,7,8),
           (0,3,6),(1,4,7),(2,5,8),
           (0,4,8),(2,4,6)]
    return any(all(board[i]==player for i in c) for c in win)

def minimax(is_max):
    if check_winner('O'): return 1
    if check_winner('X'): return -1
    if ' ' not in board: return 0

    best = -100 if is_max else 100

    for i in range(9):
        if board[i] == ' ':
            board[i] = 'O' if is_max else 'X'
            score = minimax(not is_max)
            board[i] = ' '

            if is_max:
                best = max(best, score)
            else:
                best = min(best, score)

    return best

def best_move():
    best_score = -100
    move = -1

    for i in range(9):
        if board[i] == ' ':
            board[i] = 'O'
            score = minimax(False)
            board[i] = ' '

            if score > best_score:
                best_score = score
                move = i

    return move

# Game loop
print("Positions: 0-8")
while True:
    print_board()
    pos = int(input("Enter your move: "))
    board[pos] = 'X'

    if check_winner('X'):
        print_board(); print("You Win!"); break
    if ' ' not in board:
        print("Draw!"); break

    move = best_move()
    board[move] = 'O'

    if check_winner('O'):
        print_board(); print("Computer Wins!"); break
    if ' ' not in board:
        print("Draw!"); break
```
