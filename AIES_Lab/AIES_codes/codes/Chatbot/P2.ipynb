```python
import random

def chatbot():
    print("Chatbot: Hello! Type 'exit' to quit")

    while True:
        user = input("You: ").lower()

        # Greetings
        if "hi" in user or "hello" in user:
            responses = ["Hello!", "Hi there!", "Hey!"]
            print("Chatbot:", random.choice(responses))

        # How are you
        elif "how are you" in user:
            responses = ["I'm fine!", "Doing great!", "All good!"]
            print("Chatbot:", random.choice(responses))

        # Name
        elif "name" in user:
            responses = ["I am a chatbot.", "You can call me AI bot."]
            print("Chatbot:", random.choice(responses))

        # Help
        elif "help" in user:
            print("Chatbot: I can answer basic questions!")

        # Exit
        elif "bye" in user or "exit" in user:
            print("Chatbot: Goodbye!")
            break

        # Default
        else:
            responses = ["I didn't understand.", "Can you rephrase?", "Try asking something else."]
            print("Chatbot:", random.choice(responses))


# Run
chatbot()
```
