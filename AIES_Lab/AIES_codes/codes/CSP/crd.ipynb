```python
def solutions():
    all_solutions = []

    for c in range(9, -1, -1):
        for r in range(9, -1, -1):
            for o in range(9, -1, -1):
                for s in range(9, -1, -1):
                    for a in range(9, -1, -1):
                        for d in range(9, -1, -1):
                            for n in range(9, -1, -1):
                                for g in range(9, -1, -1):
                                    for e in range(9, -1, -1):

                                        # All digits must be unique
                                        if len(set([c,r,o,s,a,d,n,g,e])) == 9:

                                            # Leading digits should not be zero
                                            if c == 0 or r == 0 or d == 0:
                                                continue

                                            cross = 10000*c + 1000*r + 100*o + 10*s + s
                                            roads = 10000*r + 1000*o + 100*a + 10*d + s
                                            danger = 100000*d + 10000*a + 1000*n + 100*g + 10*e + r

                                            if cross + roads == danger:
                                                all_solutions.append((cross, roads, danger))

    print(all_solutions)


solutions()
```
