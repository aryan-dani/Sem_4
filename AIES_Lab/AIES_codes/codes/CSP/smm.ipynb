```python
from itertools import permutations

def solution():
    all_solutions =[]
    digits = range(10)

    for perm in permutations(digits,8):
        s,e,n,d,m,o,r,y=perm

        if s == 0 or m == 0:
            continue

        send = s*1000 + e*100 + n*10+d
        more = m*1000 + o*100 + r*10+e
        money = m*10000 + o*1000 + n*100 + e*10 + y

        if(send+more == money):
            print(f"Send: {send}")
            print(f"More: {more}")
            print(f"Money: {money}")

solution()
```
