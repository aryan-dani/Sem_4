```python

# Goal state positions for fast lookup
goal_positions = {}

# Manhattan Distance
def heuristic(state, goal):
    count = 0
    for i in range(9):
        if state[i] != 0 and state[i] != goal[i]:
            count += 1
    return count

# Generate neighbors
def get_neighbors(state):
    neighbors = []
    blank = state.index(0)

    moves = {
        "UP": -3,
        "DOWN": 3,
        "LEFT": -1,
        "RIGHT": 1
    }

    for move in moves:
        new_state = state.copy()
        new_blank = blank + moves[move]

        if move == "LEFT" and blank % 3 == 0:
            continue
        if move == "RIGHT" and blank % 3 == 2:
            continue
        if new_blank < 0 or new_blank >= 9:
            continue

        new_state[blank], new_state[new_blank] = new_state[new_blank], new_state[blank]
        neighbors.append(new_state)

    return neighbors


# A* Algorithm
def a_star(initial, goal):
    open_list = []
    heapq.heappush(open_list, (0, initial))

    g_cost = {tuple(initial): 0}
    visited = set()

    while open_list:
        f, current = heapq.heappop(open_list)

        print("Current State:", current)

        if current == goal:
            print("Goal Reached!")
            return

        visited.add(tuple(current))

        for neighbor in get_neighbors(current):
            neighbor_tuple = tuple(neighbor)

            if neighbor_tuple in visited:
                continue

            g = g_cost[tuple(current)] + 1
            h = heuristic(neighbor, goal)
            f = g + h

            if neighbor_tuple not in g_cost or g < g_cost[neighbor_tuple]:
                g_cost[neighbor_tuple] = g
                heapq.heappush(open_list, (f, neighbor))


# ---- INPUT ----
initial = list(map(int, input("Enter initial state: ").split()))
goal = list(map(int, input("Enter goal state: ").split()))

# Precompute goal positions
for i in range(9):
    goal_positions[goal[i]] = i

a_star(initial, goal)
```
