```python
def hamming_distance(initial, goal):
    count = 0
    for i in range(9):
        if initial[i] != 0 and initial[i] != goal[i]:
            count += 1
    return count
```

```python
def manhattan_distance(initial, goal):
    distance = 0

    for i in range(9):
        if initial[i] != 0:
            current_value = initial[i]

            # current position
            x1, y1 = divmod(i, 3)

            # goal position
            goal_index = goal.index(current_value)
            x2, y2 = divmod(goal_index, 3)

            distance += abs(x1 - x2) + abs(y1 - y2)

    return distance


# ---- INPUT ----
initial = list(map(int, input("Enter initial state: ").split()))
goal = list(map(int, input("Enter goal state: ").split()))

print("Hamming Distance:", hamming_distance(initial, goal))
print("Manhattan Distance:", manhattan_distance(initial, goal))
```
