```python
import random

def cost(route, dist):
    return sum(dist[route[i]][route[(i+1)%len(route)]] for i in range(len(route)))

def tsp_hill_climb(dist, n):
    route = list(range(n))
    random.shuffle(route)

    while True:
        best = route
        best_cost = cost(route, dist)

        for i in range(n):
            for j in range(i+1, n):
                new_route = route[:]
                new_route[i], new_route[j] = new_route[j], new_route[i]

                new_cost = cost(new_route, dist)
                if new_cost < best_cost:
                    best = new_route
                    best_cost = new_cost

        if best == route:
            break
        route = best

    return route, best_cost


# INPUT
n = int(input())
dist = [list(map(int, input().split())) for _ in range(n)]

path, c = tsp_hill_climb(dist, n)

print("Path:", path)
print("Cost:", c)
```
