```python
import random

# Calculate total distance of a path
def calculate_cost(path, dist):
    cost = 0
    for i in range(len(path) - 1):
        cost += dist[path[i]][path[i+1]]
    cost += dist[path[-1]][path[0]]  # return to start
    return cost
```

```python
# Generate neighbor by swapping two cities
def get_neighbor(path):
    new_path = path.copy()
    i, j = random.sample(range(len(path)), 2)
    new_path[i], new_path[j] = new_path[j], new_path[i]
    return new_path
```

```python
# Hill Climbing Algorithm
def hill_climbing(dist, cities):
    current = list(range(cities))
    random.shuffle(current)

    current_cost = calculate_cost(current, dist)

    while True:
        neighbor = get_neighbor(current)
        neighbor_cost = calculate_cost(neighbor, dist)
        if neighbor_cost < current_cost:
            current = neighbor
            current_cost = neighbor_cost
        else:
            break  
    return current, current_cost


n = int(input("Enter number of cities: "))

print("Enter distance matrix:")
dist = []
for _ in range(n):
    row = list(map(int, input().split()))
    dist.append(row)

path, cost = hill_climbing(dist, n)

print("Best Path:", path)
print("Minimum Cost:", cost)
```
