```python
import pandas as pd
import numpy as np



df = pd.DataFrame({
    "Name": ["Aryan", "Sobaan", "Himali", "Swayamprakash", "Mahaveer",
             "Shruti", "Srinidhi", None, None, None, None],
    "Age": [25, 30, 35, 19, 9, 32, 60, 30, 30, 30, 30],
    "City": ["Pune", "London", "Chennai", "Pune", "Mumbai",
             None, "Kerala", None, None, None, None],
    "Phone Number": [8.668205e9, 9.909312e9, 9.422335e9, 8.390561e9,
                     7.745321e9, 9.032476e9, 8.788846e9,
                     np.nan, np.nan, np.nan, np.nan],
    "Gender": ["M", "M", "F", "M", "M", "F", "F", None, None, None, None],
    "Marks": [86, 89, 56, 90, 98, 76, 90, 89, 89, 89, 89]
})

bool_arr = df.duplicated()
df_clean = df[~bool_arr]
print("Old data\n\n",df)
print("\nNew data\n\n",df_clean)
```

```python
df_onehot = pd.get_dummies(
    df,
    columns=["Gender", "City"],
    dummy_na=True
)

print("\nOne-Hot Encoded DataFrame:\n")
print(df_onehot)
```

```python
from sklearn.preprocessing import LabelEncoder
df_label = df.copy()
le = LabelEncoder()

for col in ["Name", "City", "Gender"]:
    df_label[col] = le.fit_transform(df_label[col].astype(str))

print("Label Encoded DataFrame:\n")
print(df_label)
```

```python
drop_cols = [
    "rowid",
    "kepid",
    "kepoi_name",
    "kepler_name",
    "koi_teq_err1",
    "koi_teq_err2",
    "koi_score"
]

df = df.drop(columns=drop_cols)
```

```python
df
```

```python
cat_cols = ["koi_disposition", "koi_pdisposition"]

df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])
```

```python
df
```

```python
df["target"] = df["koi_disposition"].map({
    "CONFIRMED": 1,
    "CANDIDATE": 1,
    "FALSE POSITIVE": 0
})
```

```python
num_cols = df.select_dtypes(include=["float64", "int64"]).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
df
```

```python
flag_cols = [c for c in df.columns if "fpflag" in c]
df[flag_cols] = df[flag_cols].astype(int)
```

```python
for col in num_cols:
    q1 = df[col].quantile(0.01)
    q99 = df[col].quantile(0.99)
    df = df[(df[col] >= q1) & (df[col] <= q99)]
```

```python
df = df.drop(columns=["koi_disposition", "koi_pdisposition"])
```
